```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfh = pd.read_csv('Half-ring.csv')
dfh = dfh.dropna()
experimental_column = 'Experimental deflection ?V(mm)'
dfh[experimental_column] = dfh[experimental_column].abs()
dfh
```

```python
dfq = pd.read_csv('qurter_readings.csv')
experimental_column = 'Experimental deflection ?V(mm)'
dfq[experimental_column] = dfq[experimental_column].abs()
dfq
```

```python
load_column = 'Load(in N)'
experimental_column = 'Experimental deflection ?V(mm)'
theoretical_column = 'Theriotical Deflection ?V(mm)'


def generate_plot(experimental_column,theoretical_column,load_column,df,title):
    plt.figure(figsize=(10, 6))
    plt.plot(df[load_column], df[experimental_column], label='Experimental Deflection', marker='o')
    plt.plot(df[load_column], df[theoretical_column], label='Theoretical Deflection', marker='x')
    plt.xlabel('Load (N)')
    plt.ylabel('Deflection (mm)')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
generate_plot(experimental_column,theoretical_column,load_column,dfh,'Vertical Deflection(Half Ring) vs Load')
```

```python
experimental_column2 = 'Experimental Deflection ?H(mm)'
theoretical_column2 = 'Theriotical Deflection ?H(mm)'
generate_plot(experimental_column2,theoretical_column2,load_column,dfh,'Horizontal Deflection(Half Ring) vs Load')
```

```python
generate_plot(experimental_column,theoretical_column,load_column,dfq,'Vertical Deflection(Quarter Ring) vs Load')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
y_act_expV_hr = dfh.head(10)['Experimental deflection ?V(mm)']
x_act_expV_hr = dfh.head(10)['Load(in N)']
y_1_expV_hr = np.array(dfh.head(10)['Experimental deflection ?V(mm)']).reshape((-1, 1))
x_1_expV_hr = np.array(dfh.head(10)['Load(in N)']).reshape((-1, 1))
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

class LineGenerator:
    def __init__(self, x, y, x_act, y_act, label, x_label, y_label, title):
        self.x = np.array(x).reshape(-1, 1)  # Ensure x is in the correct shape
        self.y = np.array(y)
        self.x_act = np.array(x_act)
        self.y_act = np.array(y_act)
        self.label = label
        self.x_label = x_label
        self.y_label = y_label
        self.title = title
        self.model = None
        self.slope = None
        self.intercept = None

    def fit(self):
        self.model = LinearRegression()
        self.model.fit(self.x, self.y)

        self.slope = self.model.coef_.flatten()[0]  
        self.intercept = self.model.intercept_.item()  
        print(f"Equation of the best fit line: y = {self.slope}x + {self.intercept}")

    def plot(self):
        if self.model is None:
            raise ValueError("Model is not fitted yet. Call the fit() method first.")

        plt.scatter(self.x, self.y, color='blue', label='Data Points') 
        plt.plot(self.x, self.model.predict(self.x), color='red', label='Best Fit Line') 
        plt.plot(self.x_act, self.y_act, label=self.label, marker='o', linestyle='--') 
        equation_text = f"y = {self.slope:.2f}x + {self.intercept:.2f}"
        plt.text(0.05, 0.95, equation_text, transform=plt.gca().transAxes, fontsize=10,
                 verticalalignment='top', bbox=dict(boxstyle="round", facecolor="white", alpha=0.5))

        plt.xlabel(self.x_label)
        plt.ylabel(self.y_label)
        plt.title(self.title)
        plt.legend()
        plt.show()


```

```python
label_expV_hr = "deflection vs Load(Loading)"
x_label_expV_hr = 'Load(Loading)'
y_label_expV_hr = 'Deflection Vertical(Half Ring)'
title_expV_hr = 'Best Fit line for Deflection Vertical(Half Ring) vs Load '
line_gen = LineGenerator(x_1_expV_hr, y_1_expV_hr, x_act_expV_hr, y_act_expV_hr, label_expV_hr, x_label_expV_hr, y_label_expV_hr, title_expV_hr)
line_gen.fit()
line_gen.plot()
```
